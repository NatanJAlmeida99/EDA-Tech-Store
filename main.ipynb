```python
import pandas as pd
df_vendas = pd.read_csv('vendas.csv')
```

```python
df_vendas.head()
```

```python
df_vendas.shape[0]  # Número de linhas
```

```python
df_vendas["Receita_Total"] = df_vendas["quantidade"] * df_vendas["preco_unitario"]
display(df_vendas[["produto", "Receita_Total"]])
```

```python
df_vendas.groupby("categoria")[["Receita_Total" ]].sum()
```

```python
produto_mais_vendido = df_vendas[["produto", "quantidade"]].groupby("produto").sum().sort_values(by="quantidade", ascending=False).head(1)

produto_mais_vendido
```

```python
regiao_mais_vendas = df_vendas[["regiao", "quantidade"]].groupby("regiao").sum().sort_values(by="quantidade", ascending=False).head(1)
regiao_mais_vendas
```
